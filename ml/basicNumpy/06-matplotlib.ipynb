```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x =  np.linspace(0, 10, 100)
x
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
cosy = np.cos(x)
siny = np.sin(x)
plt.plot(x, cosy)
plt.plot(x, siny)
plt.show()
```

```python
plt.plot(x, cosy)
plt.plot(x, siny, color='r')
plt.show()
```

```python
plt.plot(x, cosy)
plt.plot(x, siny, color='r', linestyle='-.')
plt.show()
```

```python
plt.plot(x, cosy)
plt.plot(x, siny, color='r', linestyle='-.')
plt.xlim(-5, 15)
plt.show()
```

```python
plt.plot(x, cosy)
plt.plot(x, siny, color='r', linestyle='-.')
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
plt.show()
```

```python
plt.plot(x, cosy)
plt.plot(x, siny, color='r', linestyle='-.')
plt.axis([-1, 11, -2, 2])
plt.show()
```

```python
plt.plot(x, cosy)
plt.plot(x, siny, color='r', linestyle='-.')
plt.xlabel("x axis")
plt.ylabel("y label")
plt.show()
```

```python
plt.plot(x, cosy, label="sin(x)")
plt.plot(x, siny, color='r', label='cos(x)', linestyle='-.')
plt.xlabel("x axis")
plt.ylabel("y label")
plt.legend()
plt.title("plt show()")
plt.show()
```

### Scaatter Plot

```python
plt.scatter(x, siny)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color='r')
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y)
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
# 不透明度
plt.scatter(x, y, alpha=0.1)
plt.show()
```
