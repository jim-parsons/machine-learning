```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def entropy(p):
    return -p * np.log(p) - (1 - p) * np.log(1 - p)
```

```python
X = np.linspace(0.01, 0.99, 200)
```

```python
plt.plot(X, entropy(X))
plt.show()
```

```python
def gini(x):
    return 2 * x - 2 * x ** 2
```

```python
plt.plot(X, gini(X))
plt.show()
```
